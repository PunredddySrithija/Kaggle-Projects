```python
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Graphs 

#BAR GRAPH

plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='brand', y='price', color='blue')
plt.title('Mobile Brands and its Prices')
plt.xlabel('brand')
plt.ylabel('price')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# PIE CHART

brand_prices = data.groupby('brand')['price'].sum()
plt.figure(figsize=(8, 8))
plt.pie(brand_prices, labels=brand_prices.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Prices by Mobile Brand')
plt.axis('equal')
plt.show()
```

```python
#BOX PLOT

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='brand', y='price', color='skyblue')
plt.title('Price Distribution by Mobile Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#LINE PLOT

plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='brand', y='price', marker='o', color='orange')

plt.title('Average Price by Mobile Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#HISTOGRAM PLOT

plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=10, kde=True, color='blue')

plt.title('Price Distribution of Mobile Phones')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()

plt.show()
```

```python
#HEATMAP

average_prices = data.groupby('brand')['price'].mean().reset_index()
heatmap_data = average_prices.pivot(index='brand', columns='price', values='price')
plt.figure(figsize=(10, 6))
sns.heatmap(data=heatmap_data, annot=True, cmap='Blues', cbar_kws={'label': 'Average Price'})
plt.title('Heatmap of Average Prices by Mobile Brand')
plt.xlabel('Price')
plt.ylabel('Brand')
plt.tight_layout()

plt.show()
```

```python
#SCATTERPLOT

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='brand', y='price', hue='brand', style='brand', s=100, color='grey')
plt.title('Scatter Plot of Mobile Prices by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
