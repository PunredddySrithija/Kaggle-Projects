```python
# Check for missing values in each dataset
print("Missing values in Data1:")
print(data1.isnull().sum())

print("Missing values in Data2:")
print(data2.isnull().sum())

print("Missing values in Data3:")
print(data3.isnull().sum())
```

```python
# Check for duplicates in each dataset
print("Duplicates in Data1:")
print(data1.duplicated().sum())

print("Duplicates in Data2:")
print(data2.duplicated().sum())

print("Duplicates in Data3:")
print(data3.duplicated().sum())
```

```python
#Since we got duplicates, we must remove them 
dup=data2.drop_duplicates(inplace=True)
print(dup)
```

```python
# Distribution of movie ratings for data1
plt.figure(figsize=(10, 6))
sns.histplot(data1['IMDb Rating'], bins=10, kde=True)
plt.title('Distribution of Movie Ratings in Data1')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#data2
plt.figure(figsize=(10, 6))
sns.histplot(data2['Duration (minutes)'], bins=20, kde=True)
plt.title('Distribution of Movie Durations in Data2')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Average rating by genre in Data3
plt.figure(figsize=(12, 6))
sns.barplot(x='Genre', y='IMDb Rating', data=data3)
plt.title('Average Rating by Genre in Data3')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)  # Rotate x labels if needed for better readability
plt.show()
```

```python
# Rating distribution by release year for data1
plt.figure(figsize=(10, 6))
sns.lineplot(x=data1['Year'], y='IMDb Rating', data=data1)
plt.title('Rating Distribution by Release Year in Data1')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Rating vs Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (minutes)', y='IMDb Rating', data=data2)
plt.title('Rating vs Duration in Data2')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
```

```python
# Scatter plot of rating vs duration
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (minutes)', y='IMDb Rating', data=data3)
plt.title('Rating vs Duration in Data3')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
```
